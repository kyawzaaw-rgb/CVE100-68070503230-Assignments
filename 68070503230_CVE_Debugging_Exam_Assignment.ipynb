```python
def average_by_key(rows):
    total = 0
    if not rows:
        return 0
    for r in rows:
        total += r["score"]
    return total / len(rows)

print(average_by_key([{"score": 5}, {"score": 7}, {"score": 6}]))  # Expected Output: 6.0
print(average_by_key([]))   # Expected Output: 0
```

```python
def even_numbers(nums):
    evens=[]
    for n in nums:
        if n % 2 == 0:
            evens.append(n)
    return evens

print(even_numbers([1,2,3,4,5,6])) # Expected Output: [2, 4, 6]
```
